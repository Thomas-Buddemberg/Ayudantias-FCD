### Mostrar datos filtrados (>=)

```python
# Nos filtra los pokemon que tienen un HP mayor a 150
# .iloc[:,] nos muestra todas las filas
# .sort_values() nos ordena los valores segun la columna que le damos 
# (ascending=False de mayor a menor)
# (ascending=True de menor a mayor)

df[df["HP"] >= 150].iloc[:,[0, 11, 12, 13, 14, 15, 16, 17]].sort_values("HP", ascending=False)
```

#### .format y round()

```python
""" Como podemos ver, al utilizar .dropna() nos elimino 681 datos, un 75.835 % de los datos. 
Es muchisima info perdida, por lo que no es un metodo muy recomendable, solo para casos especificos """

n = 217 - 898

print("217 rows - 898 rows =", n, "rows")
print(round(((-n / 898) * 100), 3), "%")
# Formato para evitar espacio de separacion (mejor metodo)
print("{:.3f}%".format((-n / 898) * 100))
```

#### def() [funciones]

```python
def porcentaje_en_base_al_total_de_puntos(Total):
    Porcentaje = round(((Total / df["Total"].max()) * 100), 3)
    return Porcentaje
```

#### .apply(def) [aplicar funcion]

```python
# Con .apply podemos aplicar una funcion a un dataframe o columna en especifico

# Nos muestra el porcentaje del total de puntos respecto al maximo de total de puntos en el dataframe
df["Porcentaje_del_maximo_total_de_puntos"] = df["Total"].apply(porcentaje_en_base_al_total_de_puntos)
df.head(9)
```
