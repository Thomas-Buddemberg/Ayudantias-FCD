```python
import pandas as pd
```

```python
def csv_2_excel(path_csv, path_excel, nombre_csv):
    new_path_csv = path_csv.replace("\\", "/")
    new_path_excel = path_excel.replace("\\", "/")
    path_csv = new_path_csv + "/"
    path_excel = new_path_excel + "/"
    df = pd.read_csv(path_csv + "{}.csv".format(nombre_csv))
    df.to_excel(path_excel + "{}.xlsx".format(nombre_csv), index=False)
    return df
```
