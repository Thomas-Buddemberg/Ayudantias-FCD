### Differences between loc and iloc
The main distinction between **loc** and **iloc** is:

- `.loc` is label-based, which means that you have to specify rows and columns based on their row and column labels.
- `.iloc` is integer position-based, so you have to specify rows and columns by their integer position values (0-based integer position).

```python
for i in range(len(df)):
    
    if df['winner'].isna().iloc[i] == True:
        
        if df['home_score'].iloc[i] > df['away_score'].iloc[i]:
            df['winner'].iloc[i] = df['home_team'].iloc[i]
            
        elif df['home_score'].iloc[i] <  df['away_score'].iloc[i]:
            df['winner'].iloc[i] = df['away_team'].iloc[i]
            
        elif df['home_score'].iloc[i] == df['away_score'].iloc[i]:
            df['winner'].iloc[i] = 'Tie'
    
    if df['loser'].isna().iloc[i] == True:
        
        if df['home_score'].iloc[i] > df['away_score'].iloc[i]:
            df['loser'].iloc[i] = df['away_team'].iloc[i] 
            
        elif df['home_score'].iloc[i] <  df['away_score'].iloc[i]:
            df['loser'].iloc[i] = df['home_team'].iloc[i]
            
        elif df['home_score'].iloc[i] == df['away_score'].iloc[i]:
            df['loser'].iloc[i] = 'Tie'
            
df
```
